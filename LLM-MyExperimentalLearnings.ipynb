```python
#Step 2  Create embeddings + store in Vector DB

!pip uninstall numpy -y
! pip install numpy==1.26.4

!pip uninstall opencv-python -y
! pip install "opencv-python<4.10"

from sentence_transformers import SentenceTransformer
import chromadb

model = SentenceTransformer("all-mpnet-base-v2")
db = chromadb.Client()

collection = db.create_collection("knowledge_base")

docs = [
    "RV University focuses on experiential learning.",
    "Machine learning involves supervised and unsupervised learning.",
    "RISC-V is an open standard instruction set architecture."
]

embeddings = model.encode(docs).tolist()

for i, doc in enumerate(docs):
    collection.add(documents=[doc], embeddings=[embeddings[i]], ids=[str(i)])
```
